```python
# import necessary libraries for data analysis 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
import plotly.offline as pyo
# warnings.filterwarnings('ignore')
```

```python
#data creation 
num_rows = 500
probabilities = [0.4,0.4,0.2]
data = {
    'age':np.random.randint(14,65,num_rows),
    'gender':np.random.choice(['Male','Female','non-binary'],num_rows,p=probabilities),
    'demographic':np.random.choice(['urban', 'suburban','rural'],num_rows),
    'interest':np.random.choice(['Technology', 'Sports', 'Fashion', 'Travel','Porn'], num_rows),
    'device_type':np.random.choice(['Mobile','Desktop','Table'], num_rows),
    'device_brand': np.random.choice(['GooglePexel','Iphone','Tecno','Samsung','Itel','Microsoft','Vivo', 'Apple'], num_rows),
    'location':np.random.choice(['UK','Tanzania','US','Canada'], num_rows),
    'platform': np.random.choice(['Facebook', 'Instagram', 'Twitter', 'LinkedIn'], num_rows),
    'profession': np.random.choice(['Student', 'Engineer', 'Doctor', 'Teacher', 'Manager'], num_rows), 
    'time-spent': np.random.randint(1,60, num_rows),
    'income': np.random.randint(20000, 150000, num_rows),  
    'indebt': np.random.choice([True, False], num_rows), 
    'homeowner': np.random.choice([True, False], num_rows), 
    'owns_cars': np.random.choice([True, False], num_rows)  
   
}
file_path = 'data/data.csv'
df = pd.DataFrame(data)
df.to_csv(file_path, index=False)
print("CSV file is saved successfuly at:", file_path)
```

```python
#I GO DO THE DATA PREPROCESSING 
# 1. CHECKING THE DATA SHAPE
# 2. CHECKING THE NULL VALUES
# 3. SOME STATISTICAL INFORMATION
# 4. INFORMATION ABOUT THE DATA
```

```python
def checking_data_information(data_frame):
    shape = data_frame.shape
    print('*' * 80)
    
    # Calculate null values
    null_values_info = data_frame.isna().sum() / len(data_frame)
    print(null_values_info)
    print('$' * 80)
    
    # Print DataFrame information
    print(data_frame.info())
    print('&' * 80)
    
    # Calculate and print the number of duplicated values
    duplicated_value = data_frame.duplicated().sum()  # Corrected from data.duplicate() to data_frame.duplicated()
    print(f'The dataset contains {duplicated_value} duplicated values')
    
    return shape, null_values_info, data_frame.info()

# Assuming 'data2' is your DataFrame
print(checking_data_information(data2))
```

```python
#Age distribution visualization
plt.figure(figsize=(10,6))
sns.distplot(data['age'], hist=True, kde=True)
plt.title("Age Distribution Table")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# Let's Visualize the Gender Percentage in the data
data2['gender'].value_counts().sort_values(ascending=False)\
.plot(kind='pie',figsize=(8,6),
     explode=[0,0,0.1],
    labels=['Male','Non-binary','Female'],
    colors=['#780000','#03045e','#e7c6ff'],
    autopct='%1.2f%%',
    shadow=True)
plt.title("Visualize the Gender percentage in the data")
plt.show()
```

```python
#lets find how many location in the data
data2['location'].value_counts().sort_values(ascending=False)\
.plot(kind='bar',figsize=(8,6), color=['red','blue','black','aqua'])
plt.title("Findi the how many location in the data")
plt.xlabel('Locations')
plt.ylabel('Count of the values')
plt.show()
```

```python
student_data=data2[data2['profession']=='Student']
location_platform_data_frame=pd.DataFrame(student_data.groupby('location',)['platform'].value_counts().sort_values(ascending=False).unstack())
# Reshaping the DataFrame for seaborn barplot
platform_counts = location_platform_data_frame.reset_index().melt(id_vars='location', var_name='platform', value_name='count')
# Plotting with seaborn
plt.figure(figsize=(8, 6))
sns.barplot(data=platform_counts, x='location', y='count', hue='platform', palette='YlGn')
# Adding labels and title
plt.xlabel('Location')
plt.ylabel('Number of Students')
plt.title('Platform Usage by Students per Location')

# Display the plot
plt.legend(title='Platform')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
profession_platform=pd.DataFrame(data2.groupby('profession')['platform'].value_counts().sort_values(ascending=False).unstack())
profession_time=pd.DataFrame(data2.groupby('profession')['time-spent'].mean())
all_information=pd.merge(profession_platform,profession_time,on='profession')
all_information.style.background_gradient(cmap='ocean')
```
