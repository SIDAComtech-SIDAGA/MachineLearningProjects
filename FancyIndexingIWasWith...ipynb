```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mean = [0,0]
cov = [
    [1,2],
    [2,5]
]
rand = np.random.RandomState(42)
X = rand.multivariate_normal(mean,cov, 100)
X.shape
```

```python
#We can visual these points as a scatter-plot
```

```python
%matplotlib inline
import seaborn as sns
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
indeces = np.random.choice(X.shape[0],20, replace=True)
```

```python
indeces
```

```python
indeces.shape
```

```python
selection = X[indeces]
```

```python
selection.shape
```

```python
# Now to see what points were selected, lets over plot large circles at the location of the selected points
```

```python
plt.scatter(X[:,0],X[:,1], alpha=0.3)
plt.scatter(selection[:,0], selection[:,1], facecolor='none', s=200);
```
