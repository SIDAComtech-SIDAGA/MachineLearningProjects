```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('2021et.csv')
```

```python
data.head()
```

```python
print(data.columns)
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data.columns)
```

```python
data['SUBJECTS']
```

```python
subjects_df = data['SUBJECTS'].str.extractall(r"(?P<Subject>\w+)\s*-\s*'(?P<Grade>\w+)'").reset_index()
```

```python
subjects_df
```

```python
#pivot the table to have sub name
subjects_pivot = subjects_df.pivot(index='level_0', columns='Subject',values='Grade')
```

```python
subjects_pivot
```

```python
# combine with the original 
data2 = data.join(subjects_pivot)
```

```python
data2
```

```python
data2.drop(columns=['SUBJECTS'])
```

```python
data2.to_csv('outputData1.csv')
```
